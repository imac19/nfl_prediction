```python
# How much does the spread make a difference? 
# For example, if you want to be on a favorite, if the spread is as small as something like -1, is there any reason
# to take the moneyline over the spread?
# And vice versa, if you want to bet on an underdog, is it worth it to grab +1 instead of just straight moneyline?

def spread_outcome(game, favorite_wins=True, covers=True, push=False):
    
    if push:
        if game.spread_cover_result == 'push':
            return 1
        else: 
            return 0
    
    if favorite_wins and covers:
        if game.game_winner == game.spread_cover_result and game.game_winner == game.favorite:
            return 1 
        else:
            return 0
    
    elif favorite_wins and not covers:
        if game.game_winner != game.spread_cover_result and game.game_winner == game.favorite:
            return 1 
        else:
            return 0
        
    elif not favorite_wins and covers:
        if game.game_winner == game.spread_cover_result and game.game_winner != game.favorite:
            return 1 
        else:
            return 0
        
    elif not favorite_wins and not covers:
        if game.game_winner != game.spread_cover_result and game.game_winner != game.favorite:
            return 1 
        else:
            return 0

base['covered_spread'] = base.apply(lambda x: x.home_spread if x.home_spread>0 else x.away_spread, axis=1)
base['favorite'] = base.apply(lambda x: 'home' if x.home_spread<=0 else 'away', axis=1)
base['favorite_wins_and_covers'] = base.apply(lambda x: spread_outcome(x, favorite_wins=True, covers=True), axis=1)
base['favorite_wins_doesnt_cover'] = base.apply(lambda x: spread_outcome(x, favorite_wins=True, covers=False), axis=1)
base['underdog_wins_and_covers'] = base.apply(lambda x: spread_outcome(x, favorite_wins=False, covers=True), axis=1)
# base['underdog_wins_doesnt_cover'] = base.apply(lambda x: spread_outcome(x, favorite_wins=False, covers=False), axis=1)
base['spread_push'] = base.apply(lambda x: spread_outcome(x, favorite_wins=True, covers=True, push=True), axis=1)
base
```

```python
df1 = base[base.game_winner != 'push']
df1 = df1.groupby(by='covered_spread').sum()[['favorite_wins_and_covers', 'favorite_wins_doesnt_cover', 'underdog_wins_and_covers', 'spread_push']]
df1['total'] = df1.apply(lambda x: x.favorite_wins_and_covers + x.favorite_wins_doesnt_cover + x.underdog_wins_and_covers + x.spread_push, axis=1)
df1['favorite_wins_and_covers_pct'] = df1.favorite_wins_and_covers/df1.total
df1['favorite_wins_doesnt_cover_pct'] = df1.favorite_wins_doesnt_cover/df1.total
df1['underdog_wins_and_covers_pct'] = df1.underdog_wins_and_covers/df1.total
df1['spread_push_pct'] = df1.spread_push/df1.total
df1
```

```python
df2 = df1.copy().reset_index()
df2 = pd.melt(df2, id_vars = ['covered_spread'],
              value_vars=['favorite_wins_and_covers_pct', 'favorite_wins_doesnt_cover_pct', 'underdog_wins_and_covers_pct', 'spread_push_pct'],
             var_name='spread_outcome', value_name='percentage')
df2
```

```python
fig = px.line(df2, x="covered_spread", y="percentage", color='spread_outcome')
fig.show()
```
