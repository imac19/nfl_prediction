```python
# Import packages

import pandas as pd 
import numpy as np 
import os
```

```python
schedule = pd.read_csv('../data/schedule_df.csv')
qbs = pd.read_csv('../data/starting_qbs.csv')
```

```python
schedule = schedule.drop('Unnamed: 0', axis=1)
schedule = schedule.drop_duplicates()
schedule.head()
```

```python
qbs.head()
```

```python
len(qbs)
```

```python
qbs['home_score'] = qbs.home_score.astype(int)
qbs['away_score'] = qbs.away_score.astype(int)
```

```python
combined = qbs.merge(schedule, right_on = ['season', 'home_full', 'away_full', 'home_score', 'away_score'], 
                    left_on = ['season', 'home', 'away', 'home_score', 'away_score'])
```

```python
len(combined)
```

```python
combined = combined.drop('Unnamed: 0', axis=1)
```

```python
combined.head()
```

```python
combined = combined[['date', 'season', 'week', 'season_type', 'home', 'away', 'home_score', 'away_score', 
                    'home_qb', 'away_qb', 'home_qb_abv', 'away_qb_abv']]
```

```python
combined.head()
```

```python
combined.to_csv('../data/schedule_final.csv')
```
